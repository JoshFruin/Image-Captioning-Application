```python
# Open data and convert urls to images

import requests
import urllib.request
import shutil
import PIL
from PIL import Image
import cv2
from cv2 import error

def url_to_jpg(i, url, file_path):

  filename = 'image-{}.jpg'.format(i)
  full_path = '{}{}'.format(file_path, filename)

  try:
    #try:
    r = requests.get(url, stream=True, timeout=5)
    if r.status_code == 200:
        with open(full_path, 'wb') as f:
            r.raw.decode_content = True
            shutil.copyfileobj(r.raw, f)
            statfile = os.stat(full_path)
            filesize = statfile.st_size
            try:
                img = Image.open(full_path) # open the image file
                img.verify() # verify that it is, in fact an image
                train.at[i, "image_url"] = filename
                print(f'{filename} Downloaded Successfully')
            except (IOError, SyntaxError, requests.exceptions.Timeout) as e:
                train.drop(i, inplace=True)
                os.remove(full_path)
                print(f'{filename} is corrupted dropping...')

  except requests.exceptions.RequestException as e:
      print("failed:", e)
      train.drop(i, inplace=True)

  return None

# Set constants
FILENAME = 'Train_Captioning.csv'
FILEPATH = '/content/Train_Data/Train_Images/'

# Read list of URLs as Pandas Dataframe
train = pd.read_csv(FILENAME)

# Save images to the Directory by iterating through the list
for i, url in enumerate(train.values):
  url_to_jpg(i, url[1], FILEPATH)
```
